Step-1 Importing Libraries

pandas is used for data manipulation and anlaysis
numpy is used for numerical operations
matplotlib.pyplot and seaborn are used for data visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#step 2 - create the data frame
```

```python
dataframe = pd.read_csv('/content/Zomato data .csv')
print(dataframe)
```

```python
dataframe
```

```python
#Convert the data type of column-rate
```

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0]
  return value

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

Types of *Restaurant*

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of restaurant")
```

```python
##conclusion -- majority of the restaurants falls in dinning category
```

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby(['listed_in(type)'])['votes'].sum()
result = pd.DataFrame({'votes' : grouped_data} )
plt.plot(result,c="green", marker="o")
plt.xlabel("Types of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

  * Conlusion-- Dining restaurant has *recieved* maximum vote*

```python
plt.hist(dataframe['rate'], bins=10)
plt.title("ratings distribution")
plt.show()
```

##  Which mode recieves maximum ratings

```python
dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order', y='rate', data=dataframe)
```

```python
##conclusion -- offline order recieved lower rating in comparison to online order
```

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns = 'online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

# Conlusion -- Dining restaurants primarily accept offline orders, whereas cafes primarily recieve online orders. The suggests that client prefers order in person at restaurants, but prefer online ordering at cafes.
